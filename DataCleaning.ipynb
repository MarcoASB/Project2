```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime, date, time, timedelta

# File to Load (Remember to change these)
DB = "Orders Ago-Sept.csv"
DBTienda = "20190401 Base de tiendas WM y SA.csv"
Orders = pd.read_csv("Orders Ago-Sept.csv")
Stores = pd.read_csv("20190401 Base de tiendas WM y SA.csv")
```

```python
Stores['Long'] = Stores['Long'].str.replace('([+,])', '', regex = True).astype('float')
Stores['Lat'] = Stores['Lat'].str.replace('([+,])', '', regex = True).astype('float')
```

```python
Stores.head()
```

```python
Stores.dtypes
```

```python
Stores['Long'] = Stores['Long']/1000000
```

```python
Stores.head()
```

```python
Stores['Lat'] = Stores['Lat']/1000000
```

```python
Stores.head()
```

```python
Stores.to_json(r'Stores.json',orient='split')
```

```python
 db = Orders.merge(Stores, left_on='ID Tienda', right_on='Det')
```

```python
db.head(1300000)
```

```python
db1 = db.loc[:, ['folio', 'id_tipo', 'Estatus', 'ID Tienda', 'Nombre tienda', 'Region', 'Estado', 'Ciudad','Fin de captura', 'Fecha de entrega', 'Hora de entrega','Dia Sem', 'Mes', 'Venta Neta', 'Código Postal' , 'latitude', 'longitude','Lat', 'Long', 'Con servicio?']]
```

```python
db1['latitude'] = np.where(db1['latitude'].isnull(), db1['Lat'], db1['latitude'])
```

```python
db1['longitude'] = np.where(db1['longitude'].isnull(), db1['Long'], db1['longitude'])
```

```python
db1.head()
```

```python
OrdersF = db1.drop(['Lat','Long'],axis = 1)
```

```python
OrdersF.head()
```

```python
OrdersF.to_json(r'OrdersF.json',orient='split')
```

```python
# new data frame with split value columns
new = OrdersF['Fin de captura'].str.split(' ', n = 1, expand = True)
# making separate first name column from new data frame
OrdersF ['Fecha Captura']= new[0]
# making separate last name column from new data frame
OrdersF ['Hora Captura']= new[1]
# Dropping old Name columns
OrdersF.drop(columns =['Fin de captura'], inplace = True)
```
