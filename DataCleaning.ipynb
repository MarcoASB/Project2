```python
Stores['Long'] = Stores['Long'].str.replace('([+,])', '', regex = True).astype('float')
Stores['Lat'] = Stores['Lat'].str.replace('([+,])', '', regex = True).astype('float')
```

```python
Stores['Long'] = Stores['Long']/1000000
```

```python
Stores['Lat'] = Stores['Lat']/1000000
```

```python
Stores = Stores.rename(columns={'Det': 'id_tienda', 'Nombre': 'nombre_tienda', 'Region': 'region', 'Estado': 'estado', 'Ciudad': 'ciudad','Municipio': 'municipio','Proto':'proto','Long':'long','Lat': 'lat','Con servicio?': 'con_serv'})
```

```python
 db = Orders.merge(Stores, left_on='ID Tienda', right_on='id_tienda')
```

```python
db.head(1300000)
```

```python
db1 = db.loc[:, ['folio', 'id_tipo', 'Estatus', 'ID Tienda', 'Nombre tienda', 'region', 'estado', 'ciudad','Fin de captura', 'Fecha de entrega', 'Hora de entrega','Dia Sem', 'Mes', 'Venta Neta', 'Código Postal' , 'latitude', 'longitude','lat', 'long', 'con_serv']]
```

```python
db1['latitude'] = np.where(db1['latitude'].isnull(), db1['lat'], db1['latitude'])
```

```python
db1['longitude'] = np.where(db1['longitude'].isnull(), db1['long'], db1['longitude'])
```

```python
db1.head()
```

```python
OrdersF = db1.drop(['lat','long'],axis = 1)
```

```python
OrdersF = OrdersF.rename(columns={'Estatus': 'estatus', 'ID Tienda':'id_tienda' , 'Nombre tienda': 'nombre_tienda', 'Region': 'region', 'Estado': 'estado', 'Ciudad': 'ciudad','Fin de captura': 'fin_de_cap', 'Fecha de entrega': 'fecha_entr', 'Hora de entrega': 'hora_entr','Dia Sem': 'dia_sem', 'Mes':'mes', 'Venta Neta': 'venta_neta', 'Código Postal':'cp','Con servicio?': 'con_serv'})
```
