```python
datos_orders = datos_orders.rename(columns = {"ID Tienda":"Det"})
```

```python
datos = pd.merge(left=datos_stores, right=datos_orders, on="Det", how="inner")
datos
```

```python
from math import sin, cos, sqrt, atan2, radians
```

```python
datos.dtypes
```

```python
datos['latitude']=pd.to_numeric(datos['latitude'],errors='coerce')
datos['longitude']=pd.to_numeric(datos['longitude'],errors='coerce')
```

```python
datos.dtypes
```

```python
datos_dropna=datos.dropna()
datos_dropna
```

```python
datos_dropna['X']=datos_dropna['X']/1000000
datos_dropna['Y']=datos_dropna['Y']/1000000
```

```python
datos_dropna
```

```python
datos_dropna["tienda_lat"]=(np.radians(datos_dropna["Y"]))
datos_dropna["tienda_long"]=(np.radians(datos_dropna["X"]))
datos_dropna["client_lat"]=(np.radians(datos_dropna["latitude"]))
datos_dropna["client_long"]=(np.radians(datos_dropna["longitude"]))
datos_dropna
```

```python
#import math
#from math import sin, cos, sqrt, atan2, radians

#datos_dropna["distance"]=sqrt(((datos_dropna["latitude"]-datos_dropna["Y"])**2)+((datos_dropna["longitude"]-datos_dropna["X"])**2)) 
```

```python
from math import sin, cos, sqrt, atan2, radians

datos_dropna['dlon'] = datos_dropna['client_long']-datos_dropna['tienda_long']
datos_dropna['dlat'] = datos_dropna['client_lat']-datos_dropna['tienda_lat']
datos_dropna.head()
```

```python
datos_dropna.dtypes
```

```python
import math
from math import sin, cos, sqrt, atan2, radians

datos_dropna['a'] = np.sin(datos_dropna['dlat']/ 2)**2 + np.cos(datos_dropna['client_lat']) * np.cos(datos_dropna['tienda_lat'])* np.sin(datos_dropna['dlon']/ 2)**2
datos_dropna['c'] = 2 * np.arctan2(np.sqrt(datos_dropna['a']), np.sqrt(1 - datos_dropna['a']))

datos_dropna['distance_Km'] = 6373 * datos_dropna['c']
```

```python
datos_dropna
```

```python
datos_dropna.describe()
```

```python
datos_distancia_filt=datos_dropna[datos_dropna.distance_Km<100]
datos_distancia_filt
```

```python
datos_distancia_filt.describe()
```

```python
distancia_tienda = pd.pivot_table(datos_distancia_filt, values=['distance_Km','Venta Neta'], index=['Det', 'Nombre'],
                        aggfunc=np.mean)
dist_tienda=distancia_tienda.sort_values(by='distance_Km', ascending=False)
dist_tienda = dist_tienda.rename(columns = {"Venta Neta":"Venta_Promedio"})
dist_tienda
```

```python
import statsmodels.api as sm

y = dist_tienda.Venta_Promedio  # response
X = dist_tienda.distance_Km  # predictor
X = sm.add_constant(X)  # Adds a constant term to the predictor

est=sm.OLS(y, X)
est = est.fit()
est.summary()
```

```python
import matplotlib.pyplot as plt
 
var_x=dist_tienda['distance_Km']
var_y=dist_tienda['Venta_Neta']
 
plt.scatter(var_x,var_y,c='b',marker='o')
plt.line()
plt.xlabel('Distancia Pedido', fontsize=16)
plt.ylabel('Ticket Promedio', fontsize=16)
plt.title('scatter plot - Distance vs Sales',fontsize=20)
plt.show()
```
